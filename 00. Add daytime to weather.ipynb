```python
df = pd.read_csv('./assets/station1 DAYTIME.csv')
```

```python
df['Sunset'].shape
```

```python
new_is = []
for i in range (1472):
    new_i = int(str(df['Sunrise'][i])[:-2])*60 + int(str(df['Sunrise'][i])[-2:])
    new_is.append(new_i)
df['Sunrise_min'] = new_is

new_ss = []
for i in range (1472):
    new_i = int(str(df['Sunset'][i])[:-2])*60 + int(str(df['Sunset'][i])[-2:])
    new_ss.append(new_i)
df['Sunset_min'] = new_ss

df['Daytime'] = df['Sunset_min']-df['Sunrise_min']
```

```python
df.head(5).T
```

```python
df2 = pd.read_csv('./assets/station2 DAYTIME.csv')
```

```python
df2['Sunrise'] = df['Sunrise']
df2['Sunset'] = df['Sunset']
df2['Daytime'] = df['Daytime']
df2['Sunrise_min'] = df['Sunrise_min']
df2['Sunset_min'] = df['Sunset_min']
```

```python
df2.head().T
```

```python
df.to_csv('./assets/station1 DAYTIME.csv', index = False)
df2.to_csv('./assets/station2 DAYTIME.csv', index = False)
```

```python
weather_data_daytime = pd.concat([df, df2])
weather_data_daytime.to_csv('./assets/weather_data_daytime.csv', index = False)
```
